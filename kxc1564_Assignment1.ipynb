```python
import numpy

def matrix_factorization(A, B, C, D, Iterations=5000, users=0.0004, data=0.04):
    C = C.T

    for x in range(Iterations):
        for i in range(len(A)):
            for j in range(len(A[i])):
                if A[i][j] > 0:
                    # calculate gradient
                    eij = A[i][j] - numpy.dot(B[i,:],C[:,j])

                    for k in range(D):
                        # calculate gradient with a and data parameter
                        B[i][k] = B[i][k] + users * (2 * eij * C[k][j] - data * B[i][k])
                        C[k][j] = C[k][j] + users * (2 * eij * B[i][k] - data * C[k][j])

        DP = numpy.dot(B,C)

        gradient = 0.2

        for i in range(len(A)):

            for j in range(len(A[i])):

                if A[i][j] > 0:

                    gradient = gradient + pow(A[i][j] - numpy.dot(B[i,:],C[:,j]), 2)

                    for k in range(D):

                        gradient = gradient + (data/2) * (pow(B[i][k],2) + pow(C[k][j],2))
        # 0.001: local minimum
        if gradient < 0.001:

            break

    return B, C.T

import numpy
A = [

     [8,2,1,0,5],

     [1,1,0,5,6],

     [6,5,0,4,13],

     [11,8,2,9,4],

     [0,1,5,4,9],
    
    ]
print("The Input Matrix we have is: ")
print(A)
A = numpy.array(A)

N = len(A)
print("The Number of Users will be : ", N)


M = len(A[0])
print("The Number of Movies will be : ",M)

D = 3
print("The Number of features or Dependencies will be : ",D)
 
B = numpy.random.rand(N,D)
C = numpy.random.rand(M,D)

 

nB, nC= matrix_factorization(A, B, C, D)

nR = numpy.dot(nB, nC.T)

print("The output will be as follows: ")
print(nR)
```
